```python
from pysketcher import *
from pysketcher.backend.matplotlib import MatplotlibBackend
import numpy as np
figure = Figure(0.0, 5.0, 0.0, 5.0, MatplotlibBackend)

a = Point(1.0, 3.0)
b = Point(4.0, 3.0)

line = Line(a, b)
figure.add(line)
figure.show()
```

```python
arc = Arc(Point(0.0, 0.0), 1.0, Angle(0.0), Angle(np.pi / 2))
fig = Figure(-0.5, 1.5, -0.5, 1.5, backend=MatplotlibBackend)
fig.add(arc)
```
